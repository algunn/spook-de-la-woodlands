# import some importANT packages.......

```python
import numpy as np
from scipy.interpolate import interp1d
from scipy.signal import correlate
from matplotlib import pyplot as plt
import matplotlib as mpl
mpl.rcParams['figure.dpi']= 300
```

# define some definING functions........

```python
def txt_to_topo(file): # this function TAKES the .txt files the wrangler saves on the computer and GIVES numpy arrays
    
    # the wrangler .txt file is a timestamp row then 1280 rows and 4 columns per frame
    # these .txt files can be MANY frames written in succession, so lots of DATA
    # here we load that .txt file in an easily digestable way for your precious RAM
    
    # imagine the laser sheet being cast out by the wranlger in RAYS...
    # each ROW in the wrangler .txt file is a RAY corresponding to a pixel's width in the camera sensor
    # the zeroth ELEMENT in a given row is the X coordinate the laser sheet is interrupted for that RAY
    # the first ELEMENT is the Z coordinate which always zero for our wrangler, since the sheet is 2D, FORGET ABOUT IT
    # the second ELEMENT is the Y coordinate the laser sheet is interrupted for that RAY
    # the third ELEMENT is a measure of intensity of the reflection of the laser light on the obstruction - i do not know the physics/units of this quantity and have never used it
    
    # we have framelength-many RAYS
    framelength = 1280 # you can change this number if you ever fiddle with the wrangler so it doesn't produce 1280 resolution spanwise

    # a priori we don't know how many frames there are in a .txt file so append to list
    # (best practice is to NEVER append to a list instead of writing to pre-created array IF you know what length it should be a priori! because it is SLOW, and it gets EVER SLOWER)
    rawdata = [] # one array for frame data
    rawtime = [] # another for the times of the frames
    with open(file) as f: # use this method to open LARGE files, since it doesn't load the entire thing all at once
        i = 0 # use this iteratable to see if we're on a RAY or a timestamp
        for line in f: # read each line in the .txt one at a time, instead of all at once - be nice to your RAM, it never did you no harm
            if i%(framelength+1)==0: # if the remainder of the line number divide by the number of rays + another row for the timestamp is ZERO, then we're on a timestamp row
                rawtime.append(line[3:12]) # append the time stamp (in native miliseconds) to rawtime list
            else: # else, we're on a ray row
                rawdata.append(line[:-1]) # append the ray to the rawdata list
            i += 1 # update the iterable

    time = np.asarray(rawtime).astype(float)/10**3 # redefine the rawtime list as a numpy array and convert to SECONDS from miliseconds
    data = np.empty([framelength,int(np.shape(rawdata)[0]/framelength),3]) # DEFINE an empty numpy array we will load the frame data into, it has dimensions of [RAYS,FRAMES,[X,Y,INTENSITY]]

    for i in np.arange(0,np.shape(rawdata)[0]): # for each ray     
        data[i%framelength][int(i/framelength)][0] = np.float(rawdata[i].split(" ", 3)[0])/10**3 # load the X coordinate of the ray obstruction from the frame into data, convert to METERS from native milimeters
        data[i%framelength][int(i/framelength)][1] = np.float(rawdata[i].split(" ", 3)[2])/10**3 # load the Y coordinate of the ray obstruction from the frame into data, convert to METERS from native milimeters 
        data[i%framelength][int(i/framelength)][2] = np.float(rawdata[i].split(" ", 3)[3]) # load in the obstruction intensity too

    data[data[:,:,1]==0,:] = np.nan # IF the Y coordinate is ZERO, then there was NO obstruction or it spuriously wasn't registered, so make all the elements of that ray in that frame NANs
    
    data = np.squeeze(data) # IF there was only ONE frame in the .txt file, then squeeze the frame array (data) so it doesn't have a redundant FRAMES dimension in the middle
    
    return time,data # return the time and data numpy arrays
```

# surface 1

## grab data

```python
# grab the data for surface 1 (the written gravestone) and take a look

subdir = 'incriminations-of-the-cemetery-stwrangler/'
framelength = 1280 # number of pixel column rays in image
intensitymax = 1e+3 # max value that the intensity can attain
heights = np.asarray([9,10,11,12.5,13,14.5,20,24,26.5,29,36.5])*1e-2 # heights from the base of the gravestone that measurements were taken
Nrepeats = 5 # number of repeat measurements at each height
x = np.empty([len(heights),Nrepeats,framelength]) # where x data will be written
y = np.empty([len(heights),Nrepeats,framelength]) # where y data will be written
intensities = np.empty([len(heights),Nrepeats,2]) # where mean and standard deviation in intensity will be written

for i in range(len(heights)):
    for j in range(Nrepeats): # loop through heights and repeat measurements
        file = 'surface1_%.1fcm_%d.txt'%((heights[i]*1e+2),(j+1)) # get the file name (naming convention chosen when wrangler data is created in the field)
        _,data = txt_to_topo(subdir+file) # load the data from the wrangler (because we don't care about time here we don't worry about that output)
        x[i,j] = data[:,0] # save x into correct array
        y[i,j] = data[:,1] # and y...
        
        intensities[i,j] = [np.nanmean(data[:,2])/intensitymax,np.nanstd(data[:,2])/intensitymax] # get the mean and standard deviation of the reflectance intensity for the profiles, note that nans are excluded
```

## analyse data

we hypothesized that weathering is diffusive, and that weather increased up the gravestone. this would mean that the corners of the lettering get softer over time on gravestones.

one way to test this is to see how correlated a rectangular shape, with the approximate width of the lettering, is with the scan.

another clear vertical gradient in the gravestone was the visible color. does the surface also reflect the laser light differently with height?

```python
font_thickness = 5e-3 # approximate width of lettering was 5 mm
maxcorr = np.empty([len(heights),Nrepeats]) # array to store the correlation data

# to do the correlation, our data points must be equally spaced in the x-direction
# we make a new common equally-spaced x-direction array, that we will fit the data to
xstep = 1e-5
xgrid = np.arange(0,np.nanmin(x[:,:,-1]-x[:,:,0]),xstep) # this grid goes from 0 to the width of the narrowest scan, this way all regridded data will be the same length
ygrid = np.empty([len(heights),Nrepeats,len(xgrid)]) # create array to store the regridded y data

for i in range(len(heights)):
    for j in range(Nrepeats): # loop through scans
        
        # in general you can't do this with wrangler data unless the surface is 'well-behaved', always check this regridding behaves the way you expect!
        f = interp1d(x[i,j]-x[i,j][0],y[i,j],bounds_error=False,fill_value=np.nan) # create a function using scipy that will linearly interpolate y values onto a new x grid
        ygrid[i,j] = f(xgrid) # store the new gridded y data
        
        # to do the correlation, we must renormalize our data
        # we make it so the approximate x-plane of the gravestone matches the x-plane of the wrangler
        # do this by detrending the data linearly
        linfit = np.polyfit(xgrid,ygrid[i,j],deg=1) # find the straight line of best fit
        ygridtemp = ygrid[i,j]-(linfit[0]*xgrid+linfit[1]) # subtract that off of the data so on average there is no change in x within a scan
        # we also want the correlation to give comparable values between scan, and values between 0 and 1
        # renormalize the scan so it lies between 0 and 1 in the y-dimension
        ygridtemp = (ygridtemp-np.min(ygridtemp))/(np.max(ygridtemp)-np.min(ygridtemp))
        # now do the cross-correlation with a list of 1s that has the width of the lettering
        corr = correlate(ygridtemp, np.ones(int(font_thickness/xstep)), mode='valid')/int(font_thickness/xstep)
        # save the maximum of this cross-correlation for a specific scan
        maxcorr[i,j] = np.max(corr)
# maxcorr will 1 for a scan if there is a perfectly flat deepest section of the scan that has width equal to or greater than font_thickness
```

```python
lw = 0.6
alpha = 0.1
fs = 8
smark = 1
tempfigy = 0.38

fig = plt.gcf()

###################
ax0 = fig.add_axes([0.16, 0.12, 0.166, 0.83])

colors = [mpl.cm.viridis_r(x) for x in (heights-np.min(heights))/(np.max(heights)-np.min(heights))]

for i in range(len(heights)):
    for j in range(Nrepeats):
        ax0.plot(x[i,j],y[i,j]+heights[i],lw=lw,c=colors[i])
ax0.axis('equal')
ax0.set_ylim(0.15,0.45)
ax0.set_xlim(-0.03,0.03)

ax0.set_xlabel('$x$ (m)',fontsize=fs)
ax0.set_ylabel('$y+z$ (m)',fontsize=fs)
ax0.set_xticks([-0.03,0,0.03])
ax0.set_xticklabels(['-0.03','0','0.03'],fontsize=fs)
ax0.set_yticks([0.15,0.3,0.45])
ax0.set_yticklabels(['0.15','0.3','0.45'],fontsize=fs)

###################
ax1 = fig.add_axes([0.55, 0.95-tempfigy, 0.4, tempfigy])

for i in range(len(heights)):
    for j in range(Nrepeats):
        ax1.scatter(heights[i],intensities[i,j,0],s=smark,c=heights[i],vmin=np.min(heights),vmax=np.max(heights),cmap='viridis_r')
ax1.set_xlim(0,0.4)
ax1.set_ylim(0.1,0.25)
ax1.set_ylabel('$\\langle\\hat{I}\\rangle$',fontsize=fs)
ax1.set_xticks([])
ax1.set_xticklabels([],fontsize=fs)
ax1.set_yticks([0.1,0.15,0.2,0.25])
ax1.set_yticklabels(['0.1','0.15','0.2','0.25'],fontsize=fs)weatherin

###################
ax2 = fig.add_axes([0.55, 0.12, 0.4, tempfigy])

for i in range(len(heights)):
    for j in range(Nrepeats):
        ax2.scatter(heights[i],maxcorr[i,j],s=smark,c=heights[i],vmin=np.min(heights),vmax=np.max(heights),cmap='viridis_r')

ax2.set_xlim(0,0.4)
ax2.set_ylim(0.7,1)
ax2.set_xlabel('$z$ (m)',fontsize=fs)
ax2.set_ylabel('$\\max\{\\hat{y}\\star\\Pi\}$',fontsize=fs)
ax2.set_xticks([0,0.2,0.4])
ax2.set_xticklabels(['0','0.2','0.4'],fontsize=fs)
ax2.set_yticks([0.7,0.8,0.9,1])
ax2.set_yticklabels(['0.7','0.8','0.9','1'],fontsize=fs)

ax0.tick_params(width=lw,which='both',direction='in')
for axis in ['top','bottom','left','right']:
    ax0.spines[axis].set_linewidth(lw) 
ax1.tick_params(width=lw,which='both',direction='in')
for axis in ['top','bottom','left','right']:
    ax1.spines[axis].set_linewidth(lw) 
ax2.tick_params(width=lw,which='both',direction='in')
for axis in ['top','bottom','left','right']:
    ax2.spines[axis].set_linewidth(lw) 
    
fig.set_size_inches(86/25.4, 86/25.4, forward=True)
plt.savefig('fade-away.pdf', dpi=300)
```

# surface 2

## grab data

```python
# same as surface 1

subdir = 'incriminations-of-the-cemetery-stwrangler/'
framelength = 1280
heights = np.asarray([0,15,23])*1e-2
Nrepeats = 5
x = np.empty([len(heights),Nrepeats,framelength])
y = np.empty([len(heights),Nrepeats,framelength])

for i in range(len(heights)):
    for j in range(Nrepeats):
        file = 'surface2_%.1fcm_%d.txt'%((heights[i]*1e+2),(j+1))
        _,data = txt_to_topo(subdir+file)
        x[i,j] = data[:,0]
        y[i,j] = data[:,1]
```

## analyze data

we thought that the shapes in the surface could come from pattern formation due to weathering, or something to do with the structure of the marble, the grain size, or combinations.

to test this, we can find the power spectrum of the topography. it'll tell us how much of the variation in topography on a scan is captured in different scales.

```python
# again we must grid the data to use it for our chosen method of power spectra
xstep = 1e-5
xgrid = np.arange(0,np.nanmin(x[:,:,-1]-x[:,:,0]),xstep)
ygrid = np.empty([len(heights),Nrepeats,len(xgrid)])
power = np.empty([len(heights),Nrepeats,int(len(xgrid)/2+1)]) # create array to put the power spectra in
pgrid = np.linspace(0, 1/2/xstep, int(len(xgrid)/2+1)) # create the wavelength array

for i in range(len(heights)):
    for j in range(Nrepeats):
        f = interp1d(x[i,j]-x[i,j][0],y[i,j],bounds_error=False,fill_value=np.nan)
        ygrid[i,j] = f(xgrid)
        # power spectra can be made by finding the fourier transform, then getting the square of the magnitude of the modes
        # when using a non-periodic signal, we taper off the ends of the signal to avoid spurious power in wavelengths that are integer fractions of the signal length
        # this tapering is done with a hanning filter
        power[i,j] = np.abs(np.fft.rfft(ygrid[i,j]*np.hanning(len(ygrid[i,j]))))**2
```

## plot analysis

```python
lw = 0.6
alpha = 0.1
fs = 8
smark = 1
tempfigy = 0.38

fig = plt.gcf()

###################
ax0 = fig.add_axes([0.16, 0.12, 0.124, 0.83])

colors = [mpl.cm.viridis_r(x) for x in (heights-np.min(heights))/(np.max(heights)-np.min(heights))]

for i in range(len(heights)):
    for j in range(Nrepeats):
        ax0.plot(x[i,j],y[i,j]+heights[i],lw=lw,c=colors[i])
ax0.axis('equal')
ax0.set_ylim(0,0.4)
ax0.set_xlim(-0.03,0.03)

ax0.set_xlabel('$x$ (m)',fontsize=fs)
ax0.set_ylabel('$y+z$ (m)',fontsize=fs)
ax0.set_xticks([-0.03,0,0.03])
ax0.set_xticklabels(['-0.03','0','0.03'],fontsize=fs)
ax0.set_yticks([0.0,0.2,0.4])
ax0.set_yticklabels(['0.0','0.2','0.4'],fontsize=fs)

###################
ax1 = fig.add_axes([0.51, 0.12, 0.44, 0.7])

for i in range(len(heights)):
    for j in range(Nrepeats):
        ax1.plot(pgrid,power[i,j],lw=lw,c=colors[i])
sc = ax1.scatter(0,0,c=[0],cmap='viridis_r',vmin=np.min(heights),vmax=np.max(heights))
ax1.set_xscale('log')
ax1.set_yscale('log')
ax1.set_xlabel('$\\lambda$ (1/m)',fontsize=fs)
ax1.set_ylabel('$P(\\lambda)$ (m$^2$)',fontsize=fs)
ax1.set_xlim(pgrid[1],pgrid[-1])
ax1.set_ylim(1e-12,1e+2)
ax1.set_xticks([1e+2,1e+3,1e+4])
ax1.set_xticklabels(['$10^{2}$','$10^{3}$','$10^{4}$'],fontsize=fs)
ax1.set_yticks([1e-12,1e-8,1e-4,1e+0])
ax1.set_yticklabels(['$10^{-12}$','$10^{-8}$','$10^{-4}$','$10^{0}$'],fontsize=fs)

cbar_ax = fig.add_axes([0.51, 0.86, 0.44, 0.01])
c0 = fig.colorbar(sc, orientation='horizontal',cax=cbar_ax, ticks=[0,0.1,0.2])
cbar_ax.xaxis.set_label_position('top')
cbar_ax.xaxis.set_ticks_position('top')
c0.set_label('$z$ (m)',fontsize=fs)
c0.ax.set_xticklabels(['$0$','$0.1$','$0.2$'],fontsize=fs)


ax0.tick_params(width=lw,which='both',direction='in')
for axis in ['top','bottom','left','right']:
    ax0.spines[axis].set_linewidth(lw) 
ax1.tick_params(width=lw,which='both',direction='in')
for axis in ['top','bottom','left','right']:
    ax1.spines[axis].set_linewidth(lw) 
c0.ax.tick_params(width=0,which='both',direction='in')
c0.outline.set_linewidth(lw)

fig.set_size_inches(86/25.4, 86/25.4, forward=True)
plt.savefig('shrivel-up.pdf', dpi=300)
```
